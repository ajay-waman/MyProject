```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'overs': [1, 1, 2, 2, 3, 3],
    'non-striker': ['NonStriker1', 'NonStriker2', 'NonStriker1', 'NonStriker2', 'NonStriker1', 'NonStriker2'],
    'extra_type': ['wide', 'no-ball', 'wide', 'no-ball', 'wide', 'no-ball'],
}

tip = pd.DataFrame(data)

pivot_tip = tip.pivot_table(index='overs', columns=['non-striker', 'extra_type'], aggfunc=len, fill_value=0)

ax = pivot_tip.plot(kind='bar', stacked=True, figsize=(10, 6))


plt.xlabel('Overs')
plt.ylabel('Count')
plt.title('Stacked Bar Chart: Non-Striker and Extra Run Types in Each Over')


plt.legend(title='Non-Striker & Extra Type')


plt.show()



```

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'df' with columns 'batsman', 'bowlers', and 'runs_scored'

batters = tip['batter']
bowlers = tip['bowler']
runs_scored = tip['total_run']

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

x_pos = range(len(batters))
y_pos = range(len(bowlers))
x_pos, y_pos = [i + 0.5 for i in x_pos], [i + 0.5 for i in y_pos]
x_pos, y_pos = [i for _ in x_pos for i in x_pos], [i for i in y_pos for _ in y_pos]
z_pos = [0] * len(x_pos)
runs_scored_flat = [runs for sublist in runs_scored for runs in sublist]

dx = dy = 0.5
dz = runs_scored_flat

ax.bar3d(x_pos, y_pos, z_pos, dx, dy, dz, shade=True)

ax.set_xlabel('Batters')
ax.set_ylabel('Bowlers')
ax.set_zlabel('Runs Scored')
ax.set_title('3D Bar Plot of Runs Scored by Batters vs. Bowlers')

ax.set_xticks([i + 0.5 for i in range(len(batters))])
ax.set_xticklabels(batters)
ax.set_yticks([i + 0.5 for i in range(len(bowlers))])
ax.set_yticklabels(bowlers)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



tip=pd.read_csv("/content/IPL_Ball_by_Ball_2008_2022.csv.zip")


grouped_data = tip.groupby('fielders_involved')['kind'].value_counts().unstack(fill_value=0)

grouped_data['total'] = grouped_data.sum(axis=1)
grouped_data = grouped_data.sort_values(by='total', ascending=False)


top_N = 10
grouped_data = grouped_data.head(top_N)


plt.figure(figsize=(10, 6))
grouped_data.drop(columns='total').plot(kind='barh', stacked=True)


plt.xlabel('Count')
plt.ylabel('Fielders Involved')
plt.title('Horizontal Bar Chart: Fielders Involved vs. Kind of Dismissal')


plt.legend(title='Kind of Dismissal')


plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt



tip=pd.read_csv("/content/IPL_Ball_by_Ball_2008_2022.csv.zip")

x_column = 'overs'
y_column = 'batsman_run'
bubble_size_column = 'total_run'


x = tip[x_column]
y = tip[y_column]
bubble_sizes = tip[bubble_size_column]

plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=bubble_sizes, alpha=0.5, cmap='viridis')


plt.xlabel('Overs')
plt.ylabel('Batsman Runs')
plt.title('Bubble Plot: Overs vs. Batsman Runs')

plt.colorbar(label='Total Runs')


plt.show()
```
